```python
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer, LTChar,LTLine,LAParams
import os
from tqdm import tqdm
import json
import pickle


path="./data/train_dataset_Плантбокс/Атлас 2021.pdf"

Extract_Data=[]
kkk = 0
for page_layout in tqdm(extract_pages(path)):
    for element in page_layout:
        if isinstance(element, LTTextContainer):
            for text_line in element:
                try:
                    for character in text_line:
                        if isinstance(character, LTChar):
                            Font_size=character.size
                except:
                    pass
            Extract_Data.append([Font_size, (element.get_text())])
```

```python
texts = list()
last_index = 0
for i in range(len(Extract_Data)):
    if int(Extract_Data[i][0]) == 18:
        try:
            tmp = int(Extract_Data[i][1].strip())
        except:
            texts.append([j[1] for j in Extract_Data[last_index:i]])
            last_index = i
        print(Extract_Data[i])
texts = texts[5:-9]

fields = ['Другие виды. ',
          'Описание. ',
          'Ареал. ',
          'Экология. ',
          'Ресурсы. ',
          'Сырьё. ',
          'Фармакологические свойства. ',
          'Применение в медицине. ',
          'Возделывание. ',
          'Химический состав. ',
         'Литература'
]
```

```python
plants_base = dict()
for ll in range(len(texts)):
    tmp_plant = dict()

    indexes = []
    tmp_plant = dict()
    indexes = dict()
    for col in fields:
        indexes[col] = " ".join(" | ".join(texts[ll]).split()).find(col)

    indexes = sorted(indexes.items(), key=lambda x : x[1])

    last_index = None
    last_field = None
    for (k, ind) in indexes:
        if ind == -1:
            continue

        if last_index is not None:
            tmp_plant[last_field] = " ".join(" | ".join(texts[ll]).split())[last_index:ind]
        last_index = ind
        last_field = k

    tmp_plant[last_field] = " ".join(" | ".join(texts[ll]).split())[last_index:]
    plants_base[texts[ll][0].split('\n')[0]] = tmp_plant 
```

```python
plants_base['ЖЕНЬШЕНЬ (панакс женьшень)'] = plants_base["verus Oken, Aralia ginseng (C.A.Mey.) Baill.) "]
plants_base['ОРТОСИФОН ОСТИСТЫЙ (ортосифон '].update(plants_base['    '])
plants_base['СЕКУРИНЕГА ПОЛУКУСТАРНИКОВАЯ '] = plants_base['(секуринега ветвецветная, флюгея полукустарниковая)']
plants_base['ЛАПЧАТКА ПРЯМОСТОЯЧАЯ'] = plants_base['(лапчатка-узик, дикий калган, дубровка)']
plants_base['БОРЕЦ СЕВЕРНЫЙ'] = plants_base['(борец обыкновенный, б. высокий, аконит северный, а. большой)']

del plants_base['    ']
del plants_base[' ']
del plants_base["verus Oken, Aralia ginseng (C.A.Mey.) Baill.) "]
del plants_base['(секуринега ветвецветная, флюгея полукустарниковая)']
del plants_base['(лапчатка-узик, дикий калган, дубровка)']
del plants_base['(борец обыкновенный, б. высокий, аконит северный, а. большой)']

print(len(plants_base))

for i in plants_base.keys():
    print(i, ' ', len(plants_base[i]))
```

```python
count_f = dict()
for i in fields:
    count_f[i] = 0
for col in plants_base.values():
    for i in fields:
        if i in col.keys():
            count_f[i] += 1
count_f
```

```python
with open("plants_base.pkl", 'wb') as f:
    pickle.dump(plants_base, f)
```

```python
plants = list()
for plant in list(plants_base.keys()):
    plant = " ".join(plant.strip().split())
    mask = [i.isupper() or i == ' ' for i in plant]
    plants.append("".join([plant[i] for i in range(len(plant)) if mask[i]]).capitalize())
plants
```

```python
json_for_bert = list()
for plant in list(plants_base.values()):
    for k, v in plant.items():
        for ll in range(len(k), len(v), 512) :
            print(min(ll+512, len(v)-1) - ll)
            json_for_bert.append({'text' : v[ll:min(ll+512, len(v)-1)], 'label' : fields.index(k)})  
            
print(len(json_for_bert))
```

```python

with open("json_for_bert.json", 'w', encoding='utf8') as f:
    json.dump(json_for_bert, f)
```
